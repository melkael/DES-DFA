```python
# shift à gauche de maniere circulaire
def rotate(X):
    poidsFort = 1 if X > pow(2, 27) else 0
    X = X << 1
    X = (X & 0x0FFFFFFF) | poidsFort
    return X

def ShiftGauche(T):
    C = (T & 0xFFFFFFF0000000) >> 28
    D = T & 0x0000000FFFFFFF
    
    C = rotate(C)
    D = rotate(D)
    
    C = C << 28
    return (C | D)

def coupeEnDeux(entier):
    return (entier >> 32) & 0xFFFFFFFF, entier & 0xFFFFFFFF
```

```python
#Permute les bits de l'entree aPermuter, selon la table table, 
# le nombre d'éléments de table doit etre égal à tailleSortie

def permutation(aPermuter, table ,tailleEntree):
    res = 0
    for i in range(len(table)):
        # Le seul bit pouvant etre à 1 est celui qu'on permute à ce tour. 
        masque = 1 << (tailleEntree - table[i])
        # On décale ensuite ce bit ((bool)aPermuter & masque) à sa position post-permutation
        bitPermute = bool(aPermuter & masque) << (len(table) - i - 1)
        res |= bitPermute
    return res
```

```python
E = [
    32, 1,  2,  3,  4,  5,
    4,  5,  6,  7,  8,  9,
    8,  9,  10, 11, 12, 13,
    12, 13, 14, 15, 16, 17,
    16, 17, 18, 19, 20, 21,
    20, 21, 22, 23, 24, 25,
    24, 25, 26, 27, 28, 29,
    28, 29, 30, 31, 32, 1
]

def expansion(cle):
    return permutation(cle, E, 32)
```

```python
PC1 = [
    57, 49, 41, 33, 25, 17, 9,
    1,  58, 50, 42, 34, 26, 18,
    10, 2,  59, 51, 43, 35, 27,
    19, 11, 3,  60, 52, 44, 36,
    63, 55, 47, 39, 31, 23, 15,
    7,  62, 54, 46, 38, 30, 22,
    14, 6,  61, 53, 45, 37, 29,
    21, 13, 5,  28, 20, 12, 4
]

PC2 = [
    14, 17, 11, 24, 1,  5,
    3,  28, 15, 6,  21, 10,
    23, 19, 12, 4,  26, 8,
    16, 7,  27, 20, 13, 2,
    41, 52, 31, 37, 47, 55,
    30, 40, 51, 45, 33, 48,
    44, 49, 39, 56, 34, 53,
    46, 42, 50, 36, 29, 32
]

def keySchedule(K):
    sousCles = []
    
    nbPermutations = [1, 1, 2, 2, 2, 2, 2, 2, 1, 2, 2, 2, 2, 2, 2, 1]
    T = permutation(K, PC1, 64)

    for perm in nbPermutations:
        T = ShiftGauche(T) if perm == 1 else ShiftGauche(ShiftGauche(T))
        sousCles.append(permutation(T, PC2, 56))
    return sousCles
```

```python
P = [
    16, 7,  20, 21,
    29, 12, 28, 17,
    1,  15, 23, 26,
    5,  18, 31, 10,
    2,  8,  24, 14,
    32, 27, 3,  9,
    19, 13, 30, 6,
    22, 11, 4,  25
]

def f(R, sousCle):
    # Apres expansion, R fait 48 bits, comme la sous clé
    T = expansion(R) ^ sousCle
    
    res = 0

    for i in range(1, 9):
        # 48 bits / 8 = 8 blocs de 6 bits à traiter séparément 
        bloc = T >> (48 - i*6)
        poidsFortFoisDeux = (bloc & 0b100000) >> 4
        poidsFaible = bloc & 1
        
        ligne = poidsFortFoisDeux + poidsFaible
        colonne = (bloc & 0b011110) >> 1

        # Ensuite passage de chaque bloc dans une sbox => 4 bits en sortie
        # 4 * 8 = 32 bits en sortie au total
        res = res | (Sbox[i - 1][ligne][colonne] << (32 - i*4))
    # Permutation du résultat
    return permutation(res, P, 32)
```

```python
IP = [
    58, 50, 42, 34, 26, 18, 10, 2,
    60, 52, 44, 36, 28, 20, 12, 4,
    62, 54, 46, 38, 30, 22, 14, 6,
    64, 56, 48, 40, 32, 24, 16, 8,
    57, 49, 41, 33, 25, 17, 9,  1,
    59, 51, 43, 35, 27, 19, 11, 3,
    61, 53, 45, 37, 29, 21, 13, 5,
    63, 55, 47, 39, 31, 23, 15, 7
]
    
IPinv = [
    40, 8,  48, 16, 56, 24, 64, 32,
    39, 7,  47, 15, 55, 23, 63, 31,
    38, 6,  46, 14, 54, 22, 62, 30,
    37, 5,  45, 13, 53, 21, 61, 29,
    36, 4,  44, 12, 52, 20, 60, 28,
    35, 3,  43, 11, 51, 19, 59, 27,
    34, 2,  42, 10, 50, 18, 58, 26,
    33, 1,  41, 9,  49, 17, 57, 25
]

def DES(clair, K):
    # Dérivation de la clé K (64 bits) en 16 sous-cles
    sousCles = keySchedule(K)
    
    # Permutation initiale IP
    clairIP = permutation(clair, IP, 64)
    L, R = coupeEnDeux(clairIP)

    # 1 tour de Feistel par sous clé (16)
    for i in sousCles:
        LiPlus1 = R
        RiPlus1 = L ^ f(R, i)

        L = LiPlus1
        R = RiPlus1
    
    invChiffre = (R << 32) | L
    return permutation(invChiffre, IPinv, 64)
```

```python
# Recoit une liste de listes et trouve l'éléments qui leur est commun
def commun(liste):
    result = set(liste[0])
    for l in liste[1:]:
        result.intersection_update(l)
    return result.pop()

# génère un masque de 6 bits pour isoler les entrées de chaque sbox
def mask48(sbox):
    mask = "111111"
    zerosGauche = sbox * 6 * '0'
    mask = zerosGauche + mask
    while(len(mask) < 48):
        mask += '0'
    return int(mask, 2)

# génère un masque de 4 bits pour isoler les sorties de chaque sbox
def mask32(sbox):
    mask = "1111"
    return int(mask + "0" * (28 - sbox * 4), 2)
```

```python
def calcLigneColonneSbox(sbox, expanded, valeurATester):
    # on garde les 6 bits qui passent par la sbox
    tmp = expanded & mask48(sbox)
    # puis on enleve les tous les zeros à droite
    tmp >>= (7 - sbox) * 6
    tmp ^= valeurATester
    
    # meme facon de calculer l'entrée de la sbox que dans DES normal
    poidsFortFoisDeux = (tmp & 0b100000) >> 4
    poidsFaible = tmp & 1
    ligne = poidsFortFoisDeux + poidsFaible   # valeur entre 0 et 3
    colonne = (tmp & 0b011110) >> 1           # valeur entre 0 et 15
    return ligne, colonne

def solutionValide(sbox, L16_xor_L16f, ligne, colonne, ligneFaux, colonneFaux):
    ver = L16_xor_L16f & mask32(sbox) # on garde les 4 bits qui concernent la sbox courante
    ver >>= (7 - sbox) * 4            # on enleve les zeros de droite
    xor = Sbox[sbox][ligne][colonne] ^ Sbox[sbox][ligneFaux][colonneFaux]
    return ver == xor                 # s'il y a égalité alors on a trouvé une solution potentielle à l'équation
```

```python
def choisitFautes():
    aAtq = {'sbox{}'.format(i): [] for i in range(8)}

    for index, i in enumerate(chiffreFaux):
        xor = chiffreJuste ^ i
        xor = permutation(xor, IP, 64)
        
        L16, R16 = coupeEnDeux(xor)
        exp = bin(expansion(R16))[2:].zfill(48)
        expListe = [exp[i:i+6] for i in range(0, 48, 6)]
        
        for idx, bloc in enumerate(expListe):
            if bloc != '000000':
                aAtq['sbox{}'.format(idx)].append(index)
    return aAtq
```

```python
import pprint

Pinv = [
    9,  17, 23, 31,
    13, 28, 2,  18,
    24, 16, 30, 6,
    26, 20, 10, 1,
    8,  14, 25, 3,
    4,  29, 11, 19,
    32, 12, 22, 7,
    5,  27, 15, 21
]

def rechercheK16(chiffreJuste, chiffreFaux, verbose=False):
    K16 = 0x000000000000
    sol = {"sbox{}".format(i): [] for i in range(8)}

    aAttaquer = choisitFautes()
    
    
    for sbox, liste in aAttaquer.items():
        lChiffreFaux = []
        for i in liste:
            lChiffreFaux.append(chiffreFaux[i])
        aAttaquer[sbox] = lChiffreFaux
    
    L16, R15 = coupeEnDeux(permutation(chiffreJuste, IP, 64))
    
    # Attaque des 8 Sbox à l'aide des 8 equations établies, chaque attaque va nous permettre de connaitre 8 bits de K16
    for s in range(8):
        # Chacune des 8 Sbox va etre attaquée par 6 chiffrés faux
        for f in range(6): 
            L16f, R15f = coupeEnDeux(permutation(aAttaquer['sbox{}'.format(s)][f], IP, 64))
    
            # On calcule les éléments de l'équation
            L16_xor_L16f = permutation(L16 ^ L16f, Pinv, 32)
            E_R15 = expansion(R15) 
            E_R15f = expansion(R15f)
            
            # currSols sert a stocker la liste de solutions possibles pour l'équation et la sbox courantes
            currSols = []
            # test de toutes les valeurs possibles d'une portion de k16 (6 bits)
            for x in range(pow(2, 6)):
                
                ligne, colonne = calcLigneColonneSbox(s, E_R15, x)
                ligneFaux, colonneFaux = calcLigneColonneSbox(s, E_R15f, x)
                
                if (solutionValide(s, L16_xor_L16f, ligne, colonne, ligneFaux, colonneFaux)):
                    currSols.append(x)
            sol["sbox{}".format(s)].append(currSols)

        # la solution est l'octet qui est potentiellement solution pour toutes les fautes
        solution = commun(sol["sbox{}".format(s)])
        K16 = K16 << 6
        K16 = K16 | solution

        if verbose is True:
            pp = pprint.PrettyPrinter(indent=4)
            print("Sbox", s + 1)
            print("Solutions potentielles :")
            pp.pprint(sol['sbox{}'.format(s)])
            print("Solution", s+1 , "=", hex(solution))
            print("K16 courant =", hex(K16))
            
    return K16
```

```python
def setKthBit(n,k): 
    return ((1 << k) | n) 
```

```python
from itertools import product

PC1inv = [
    8,  16, 24, 56, 52, 44,
    36, 0,  7,  15, 23, 55,
    51, 43, 35, 0,  6,  14,
    22, 54, 50, 42, 34, 0,
    5,  13, 21, 53, 49, 41,
    33, 0,  4,  12, 20, 28,
    48, 40, 32, 0,  3,  11,
    19, 27, 47, 39, 31, 0,
    2,  10, 18, 26, 46, 38,
    30, 0,  1,  9,  17, 25,
    45, 37, 29, 0
]

# Les bits 9, 18, 22, 25, 35, 38, 43 et 54 sont perdus, ils sont donc mis à 0 dans cette table
PC2inv = [
    5,  24, 7,  16, 6,  10,
    20, 18, 0,  12, 3,  15,
    23, 1,  9,  19, 2,  0,
    14, 22, 11, 0,  13, 4,
    0,  17, 21, 8,  47, 31,
    27, 48, 35, 41, 0,  46,
    28, 0,  39, 32, 25, 44,
    0,  37, 34, 43, 29, 36,
    38, 45, 33, 26, 42, 0,
    30,  40
]

def rechercheK56b(clair, chiffre, K16):
    # On remet les 48 bits de K16 à la bonne position pour K (on "remonte" le key schedule)
    # Inverse de PC2 : 48 -> 56 bits, 8 bits faux dont on connait les positions
    # Inverse de PC1 : 56 -> 64 bits, 8 bits faux (parité, pas d'incidence sur le calcul). Par contre les 8 bits
    # qui étaient faux auparavant changent de position mais on les connait toujours
    K48b = permutation(permutation(K16, PC2inv, 48), PC1inv, 56);
    
    print("48 bons bits de K = {}".format(hex(K48b)))
    
    # Recherche exhaustive sur les 8 bits manquants de K
    # OSEF des bits de parité ici puisqu'ils n'ont pas d'incidence sur le calcul
    masques = list(product([0, 1], repeat=8))
    positionsARetrouver = [50, 49, 45, 44, 13, 10, 6, 4]
    
    for mask in masques:
        hypothese = 0
        for i in range(8):
            if mask[i] == 1:
                hypothese = setKthBit(hypothese, k=positionsARetrouver[i])
        hypothese |= K48b
        if chiffre == DES(clair, hypothese):
            return hypothese
    
    print("Impossible de retrouver K56, erreur")
    raise
```

```python
import struct

def decale(bit, offset):
    return bit << offset

def bitsParite(K56b):
    strCle = bin(K56b)[2:]
    while len(strCle) < 64:
        strCle = "0" + strCle
    cle = ''
    i = 0
    while i < 64:
        currByte = strCle[i:i+8]
        if currByte.count('1') % 2 == 0:
            currByte = currByte[:-1] + '1'
        cle += currByte
        i+=8
    return int(cle, 2)
```
